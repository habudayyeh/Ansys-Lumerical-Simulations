```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pickle
from mpl_toolkits.mplot3d import Axes3D
from sklearn import preprocessing as pre
from scipy.ndimage import gaussian_filter
from scipy import ndimage
```

```python
g = 0.05
Ex = 2.013
x = np.linspace(-0.25, 0.25, 200)
E_photon = x*np.tan(-0.70687) + 1.58575627/np.cos(-0.70687)
E_low_pol = 0.5*(E_photon + Ex - np.sqrt((E_photon - Ex)**2 + 4*g**2))
E_up_pol = 0.5*(E_photon + Ex + np.sqrt((E_photon - Ex)**2 + 4*g**2))
```

```python
plt.plot(x, Ex*np.ones_like(x))
plt.plot(x, E_photon)
plt.plot(x, E_low_pol, 'r')
plt.plot(x, E_up_pol, 'r')
```
